```python
#LLM Pattern 2
from pathlib import Path
from typing import Any, List
from tqdm import tqdm
import yaml

# -------------------- CONFIG --------------------
DATADIR = Path("/Users/ankita/Desktop/llm_prompt_used_files_5000_10000")  # change to your LLM run dataset folder

MATCHED_OUTPUT = Path("/Users/ankita/Desktop/Thesis-work/LLM_Experiment/Python matcher metrics/llm_pattern2_matched_files_python_based.txt")
NON_MATCHED_OUTPUT = Path("/Users/ankita/Desktop/Thesis-work/LLM_Experiment/Python matcher metrics/llm_pattern2_non_matched_files_python_based.txt")
ERROR_OUTPUT = Path("/Users/ankita/Desktop/Thesis-work/LLM_Experiment/Python matcher metrics/llm_pattern2_error_files_python_based.txt")

# -------------------- YAML loader: prevent 'on' -> True --------------------
class NoBoolSafeLoader(yaml.SafeLoader):
    pass

for ch, mappings in list(NoBoolSafeLoader.yaml_implicit_resolvers.items()):
    NoBoolSafeLoader.yaml_implicit_resolvers[ch] = [
        (tag, rx) for (tag, rx) in mappings
        if tag != "tag:yaml.org,2002:bool"
    ]

def safeload_nobool(text: str) -> Any:
    try:
        return yaml.load(text, Loader=NoBoolSafeLoader)
    except Exception:
        return None

# -------------------- Pattern checks --------------------
def has_top_level_name(doc: Any) -> bool:
    return isinstance(doc, dict) and "name" in doc

def has_trigger_block(doc: Any) -> bool:
    return isinstance(doc, dict) and ("on" in doc or "Triggers" in doc or "triggers" in doc)

def get_jobs(doc: Any) -> dict:
    if not isinstance(doc, dict):
        return {}
    jobs = doc.get("jobs")
    return jobs if isinstance(jobs, dict) else {}

def step_has_name_and_run(step: Any) -> bool:
    return isinstance(step, dict) and "name" in step and "run" in step

def step_has_uses(step: Any) -> bool:
    return (
        isinstance(step, dict)
        and len(step) > 0
        and list(step.keys())[0] == "uses"
    )

def job_matches(job: Any) -> bool:
    if not isinstance(job, dict):
        return False

    if "runs-on" not in job:
        return False

    steps = job.get("steps")
    if not isinstance(steps, list):
        return False

    has_named_run_step = any(step_has_name_and_run(step) for step in steps)
    has_uses_step = any(step_has_uses(step) for step in steps)

    return has_named_run_step and has_uses_step

def matches_pattern(doc: Any) -> bool:
    if not has_top_level_name(doc):
        return False

    if not has_trigger_block(doc):
        return False

    jobs = get_jobs(doc)
    if not jobs:
        return False

    return any(job_matches(job_body) for job_body in jobs.values())

# -------------------- Run over dataset folder --------------------
allpaths: List[Path] = sorted(DATADIR.rglob("*.Pipeline"))

print(f"Scanning {len(allpaths):,} .Pipeline files under {DATADIR}")

matched: List[str] = []
non_matched: List[str] = []
errors: List[str] = []

for p in tqdm(allpaths, desc="Checking"):
    try:
        raw = p.read_text(encoding="utf-8", errors="ignore")
        doc = safeload_nobool(raw)

        if doc is None:
            errors.append(p.name)
            non_matched.append(p.name)
            continue

        if matches_pattern(doc):
            matched.append(p.name)
        else:
            non_matched.append(p.name)

    except Exception:
        errors.append(p.name)
        non_matched.append(p.name)

matched_unique = sorted(set(matched), key=str.lower)
non_matched_unique = sorted(set(non_matched), key=str.lower)
errors_unique = sorted(set(errors), key=str.lower)

MATCHED_OUTPUT.write_text(
    "\n".join(matched_unique) + ("\n" if matched_unique else ""),
    encoding="utf-8"
)

NON_MATCHED_OUTPUT.write_text(
    "\n".join(non_matched_unique) + ("\n" if non_matched_unique else ""),
    encoding="utf-8"
)

ERROR_OUTPUT.write_text(
    "\n".join(errors_unique) + ("\n" if errors_unique else ""),
    encoding="utf-8"
)

print("\n=== Results ===")
print(f"Total .Pipeline files scanned : {len(allpaths):,}")
print(f"Matched files                 : {len(matched_unique):,}")
print(f"Non-matched files             : {len(non_matched_unique):,}")
print(f"Parse/error files             : {len(errors_unique):,}")

print(f"\nMatched filenames saved to    : {MATCHED_OUTPUT.resolve()}")
print(f"Non-matched filenames saved to: {NON_MATCHED_OUTPUT.resolve()}")
print(f"Error filenames saved to      : {ERROR_OUTPUT.resolve()}")
```

```python
#LLM Pattern 3
from pathlib import Path
from typing import Any, List
from tqdm import tqdm
import yaml

# -------------------- CONFIG --------------------
DATADIR = Path("/Users/ankita/Desktop/llm_prompt_used_files_1000_2000")

MATCHED_OUTPUT = Path("/Users/ankita/Desktop/Thesis-work/LLM_Experiment/Python matcher metrics/llm_pattern3_matched_files_python_based.txt")
NON_MATCHED_OUTPUT = Path("/Users/ankita/Desktop/Thesis-work/LLM_Experiment/Python matcher metrics/llm_pattern3_non_matched_files_python_based.txt")
ERROR_OUTPUT = Path("/Users/ankita/Desktop/Thesis-work/LLM_Experiment/Python matcher metrics/llm_pattern3_error_files_python_based.txt")

# -------------------- YAML loader: prevent 'on' -> True --------------------
class NoBoolSafeLoader(yaml.SafeLoader):
    pass

for ch, mappings in list(NoBoolSafeLoader.yaml_implicit_resolvers.items()):
    NoBoolSafeLoader.yaml_implicit_resolvers[ch] = [
        (tag, rx) for (tag, rx) in mappings
        if tag != "tag:yaml.org,2002:bool"
    ]

def safeload_nobool(text: str) -> Any:
    try:
        return yaml.load(text, Loader=NoBoolSafeLoader)
    except Exception:
        return None

# -------------------- Pattern checks --------------------
def has_top_level_name(doc: Any) -> bool:
    return isinstance(doc, dict) and "name" in doc

def has_push_branches_trigger(doc: Any) -> bool:
    if not isinstance(doc, dict):
        return False

    on_block = doc.get("on")
    if not isinstance(on_block, dict):
        return False

    push_block = on_block.get("push")
    if not isinstance(push_block, dict):
        return False

    return "branches" in push_block

def get_jobs(doc: Any) -> dict:
    if not isinstance(doc, dict):
        return {}

    jobs = doc.get("jobs")
    return jobs if isinstance(jobs, dict) else {}

def step_has_name_and_uses(step: Any) -> bool:
    return isinstance(step, dict) and "name" in step and "uses" in step

def step_has_name_and_run(step: Any) -> bool:
    return isinstance(step, dict) and "name" in step and "run" in step

def job_matches_ordered_steps(job: Any) -> bool:
    if not isinstance(job, dict):
        return False

    if "runs-on" not in job:
        return False

    steps = job.get("steps")
    if not isinstance(steps, list):
        return False

    found_name_uses = False

    for step in steps:
        if not found_name_uses:
            if step_has_name_and_uses(step):
                found_name_uses = True
        else:
            if step_has_name_and_run(step):
                return True

    return False

def matches_pattern(doc: Any) -> bool:
    if not has_top_level_name(doc):
        return False

    if not has_push_branches_trigger(doc):
        return False

    jobs = get_jobs(doc)
    if not jobs:
        return False

    return any(job_matches_ordered_steps(job_body) for job_body in jobs.values())

# -------------------- Run over dataset folder --------------------
allpaths: List[Path] = sorted(DATADIR.rglob("*.Pipeline"))

print(f"Scanning {len(allpaths):,} .Pipeline files under {DATADIR}")

matched: List[str] = []
non_matched: List[str] = []
errors: List[str] = []

for p in tqdm(allpaths, desc="Checking"):
    try:
        raw = p.read_text(encoding="utf-8", errors="ignore")
        doc = safeload_nobool(raw)

        if doc is None:
            errors.append(p.name)
            non_matched.append(p.name)
            continue

        if matches_pattern(doc):
            matched.append(p.name)
        else:
            non_matched.append(p.name)

    except Exception:
        errors.append(p.name)
        non_matched.append(p.name)

matched_unique = sorted(set(matched), key=str.lower)
non_matched_unique = sorted(set(non_matched), key=str.lower)
errors_unique = sorted(set(errors), key=str.lower)

MATCHED_OUTPUT.write_text(
    "\n".join(matched_unique) + ("\n" if matched_unique else ""),
    encoding="utf-8"
)

NON_MATCHED_OUTPUT.write_text(
    "\n".join(non_matched_unique) + ("\n" if non_matched_unique else ""),
    encoding="utf-8"
)

ERROR_OUTPUT.write_text(
    "\n".join(errors_unique) + ("\n" if errors_unique else ""),
    encoding="utf-8"
)

print("\n=== Results ===")
print(f"Total .Pipeline files scanned : {len(allpaths):,}")
print(f"Matched files                 : {len(matched_unique):,}")
print(f"Non-matched files             : {len(non_matched_unique):,}")
print(f"Parse/error files             : {len(errors_unique):,}")

print(f"\nMatched filenames saved to    : {MATCHED_OUTPUT.resolve()}")
print(f"Non-matched filenames saved to: {NON_MATCHED_OUTPUT.resolve()}")
print(f"Error filenames saved to      : {ERROR_OUTPUT.resolve()}")
```

```python
#LLM Pattern 4
from pathlib import Path
from typing import Any, List
from tqdm import tqdm
import yaml
import re

# -------------------- CONFIG --------------------
DATADIR = Path("/Users/ankita/Desktop/llm_prompt_used_files_20000_25000")

MATCHED_OUTPUT = Path("/Users/ankita/Desktop/Thesis-work/LLM_Experiment/Python matcher metrics/llm_pattern4_matched_files_python_based.txt")
NON_MATCHED_OUTPUT = Path("/Users/ankita/Desktop/Thesis-work/LLM_Experiment/Python matcher metrics/llm_pattern4_non_matched_files_python_based.txt")
ERROR_OUTPUT = Path("/Users/ankita/Desktop/Thesis-work/LLM_Experiment/Python matcher metrics/llm_pattern4_error_files_python_based.txt")

# -------------------- YAML loader: prevent 'on' -> True --------------------
class NoBoolSafeLoader(yaml.SafeLoader):
    pass

for ch, mappings in list(NoBoolSafeLoader.yaml_implicit_resolvers.items()):
    NoBoolSafeLoader.yaml_implicit_resolvers[ch] = [
        (tag, rx) for (tag, rx) in mappings
        if tag != "tag:yaml.org,2002:bool"
    ]

def safeload_nobool(text: str) -> Any:
    try:
        return yaml.load(text, Loader=NoBoolSafeLoader)
    except Exception:
        return None

# -------------------- Pattern checks --------------------
def norm_key(k: str) -> str:
    return re.sub(r"[\s\-_]+", "", k.strip().lower())

def has_top_level_name(doc: Any) -> bool:
    return isinstance(doc, dict) and "name" in doc

def has_ordered_push_branches_then_pull_request(doc: Any) -> bool:
    if not isinstance(doc, dict):
        return False

    on_block = doc.get("on")
    if not isinstance(on_block, dict):
        return False

    seen_push_with_branches = False

    for k, v in on_block.items():
        if not isinstance(k, str):
            continue

        nk = norm_key(k)

        if nk == "push":
            if isinstance(v, dict) and "branches" in v:
                seen_push_with_branches = True

        elif nk == "pullrequest":
            return seen_push_with_branches

    return False

def get_build_job(doc: Any) -> Any:
    if not isinstance(doc, dict):
        return None

    jobs = doc.get("jobs")
    if not isinstance(jobs, dict):
        return None

    return jobs.get("build")

def step_has_uses(step: Any) -> bool:
    return (
        isinstance(step, dict)
        and len(step) > 0
        and list(step.keys())[0] == "uses"
    )

def step_has_name_and_run(step: Any) -> bool:
    return isinstance(step, dict) and "name" in step and "run" in step

def build_job_matches_ordered_steps(job: Any) -> bool:
    if not isinstance(job, dict):
        return False

    if "runs-on" not in job:
        return False

    steps = job.get("steps")
    if not isinstance(steps, list):
        return False

    state = 0

    for step in steps:
        if state == 0 and step_has_uses(step):
            state = 1

        elif state == 1 and step_has_name_and_run(step):
            state = 2

        elif state == 2 and step_has_name_and_run(step):
            return True

    return False

def matches_pattern(doc: Any) -> bool:
    if not has_top_level_name(doc):
        return False

    if not has_ordered_push_branches_then_pull_request(doc):
        return False

    build_job = get_build_job(doc)
    if build_job is None:
        return False

    return build_job_matches_ordered_steps(build_job)

# -------------------- Run over dataset folder --------------------
allpaths: List[Path] = sorted(DATADIR.rglob("*.Pipeline"))

print(f"Scanning {len(allpaths):,} .Pipeline files under {DATADIR}")

matched: List[str] = []
non_matched: List[str] = []
errors: List[str] = []

for p in tqdm(allpaths, desc="Checking"):
    try:
        raw = p.read_text(encoding="utf-8", errors="ignore")
        doc = safeload_nobool(raw)

        if doc is None:
            errors.append(p.name)
            non_matched.append(p.name)
            continue

        if matches_pattern(doc):
            matched.append(p.name)
        else:
            non_matched.append(p.name)

    except Exception:
        errors.append(p.name)
        non_matched.append(p.name)

matched_unique = sorted(set(matched), key=str.lower)
non_matched_unique = sorted(set(non_matched), key=str.lower)
errors_unique = sorted(set(errors), key=str.lower)

MATCHED_OUTPUT.write_text(
    "\n".join(matched_unique) + ("\n" if matched_unique else ""),
    encoding="utf-8"
)

NON_MATCHED_OUTPUT.write_text(
    "\n".join(non_matched_unique) + ("\n" if non_matched_unique else ""),
    encoding="utf-8"
)

ERROR_OUTPUT.write_text(
    "\n".join(errors_unique) + ("\n" if errors_unique else ""),
    encoding="utf-8"
)

print("\n=== Results ===")
print(f"Total .Pipeline files scanned : {len(allpaths):,}")
print(f"Matched files                 : {len(matched_unique):,}")
print(f"Non-matched files             : {len(non_matched_unique):,}")
print(f"Parse/error files             : {len(errors_unique):,}")

print(f"\nMatched filenames saved to    : {MATCHED_OUTPUT.resolve()}")
print(f"Non-matched filenames saved to: {NON_MATCHED_OUTPUT.resolve()}")
print(f"Error filenames saved to      : {ERROR_OUTPUT.resolve()}")
```

```python
# Compare filenames from given 2 files containing list of filenames
# Config — edit these two paths:
file1 = "/Users/ankita/Desktop/Thesis-work/LLM_Experiment/Patternmatching_files_all_files_LLM_pattern1.txt"
file2 = "/Users/ankita/Desktop/Thesis-work/LLM_Experiment/Python matcher metrics/llm_pattern1_matched_files_python_based.txt"

# Normalization flags
CASE_SENSITIVE = False          # set True if you want exact case matching
REQUIRE_EXTENSION = ".Pipeline" # set None to allow all; else only keep names with this suffix

# Output folder (reuses SAVE_DIR if you already defined it; else uses current dir)
try:
    SAVE_DIR
except NameError:
    from pathlib import Path
    SAVE_DIR = Path(".")
SAVE_DIR.mkdir(parents=True, exist_ok=True)

from pathlib import Path

def _load_names(path, case_sensitive=False, require_ext=None):
    names = []
    path = Path(path)
    with path.open("r", encoding="utf-8", errors="ignore") as f:
        for line in f:
            s = line.strip()
            if not s or s.startswith("#"):
                continue
            base = Path(s).name  # strip directories if present
            if require_ext and not base.endswith(require_ext):
                continue
            names.append(base if case_sensitive else base.lower())
    # unique + stable order for display
    uniq = sorted(set(names))
    return uniq

list1 = _load_names(file1, CASE_SENSITIVE, REQUIRE_EXTENSION)
list2 = _load_names(file2, CASE_SENSITIVE, REQUIRE_EXTENSION)

set1, set2 = set(list1), set(list2)

matches = sorted(set1 & set2)
only1  = sorted(set1 - set2)
only2  = sorted(set2 - set1)

# Counts
total1 = len(set1)
total2 = len(set2)
match_count = len(matches)
nonmatch_count = len(only1) + len(only2)

print("=== Filename comparison summary ===")
print(f"File 1: {file1}")
print(f"File 2: {file2}")
print(f"Case sensitive: {CASE_SENSITIVE}")
print(f"Require extension: {REQUIRE_EXTENSION if REQUIRE_EXTENSION else 'None'}\n")

print(f"Total unique in file1: {total1}")
print(f"Total unique in file2: {total2}")
print(f"Matching count        : {match_count}")
print(f"Non-matching total    : {nonmatch_count}")
print(f"  ├─ Only in file1    : {len(only1)}")
print(f"  └─ Only in file2    : {len(only2)}")

# Save results
out_matches = SAVE_DIR / "compare_matches.txt"
out_only1   = SAVE_DIR / "compare_only_in_file1.txt"
out_only2   = SAVE_DIR / "compare_only_in_file2.txt"

out_matches.write_text("\n".join(matches) + ("\n" if matches else ""), encoding="utf-8")
out_only1.write_text("\n".join(only1) + ("\n" if only1 else ""), encoding="utf-8")
out_only2.write_text("\n".join(only2) + ("\n" if only2 else ""), encoding="utf-8")

print("\nSaved:")
print(f"• Matches           → {out_matches}")
print(f"• Only in file1     → {out_only1}")
print(f"• Only in file2     → {out_only2}")

# Optional: show a quick peek of differences
def _peek(lst, n=10):
    return "\n".join(lst[:n]) if lst else "(none)"

print("\nTop matches (up to 10):")
print(_peek(matches))
print("\nTop only-in-file1 (up to 10):")
print(_peek(only1))
print("\nTop only-in-file2 (up to 10):")
print(_peek(only2))
```
